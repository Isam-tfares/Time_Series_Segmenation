<h1>ClaSPy: A Python package for time series segmentation</h1>

<p>The ClaSP (Classification by Score Profile) algorithm is a method used to segment time series by identifying significant changes in the behavior of the series. The ClaSP approach is based on the idea of calculating similarity scores between segments of the time series and then using these scores to detect change points. More precisely, the ClaSP algorithm follows these main steps:</p>
<ol>
<li>Calculation of similarity scores: Divide the time series into segments and calculate similarity scores between adjacent segments. These scores can be based on different similarity measures, such as correlation, Euclidean distance, or other measures adapted to the specific context of the time series.</li>

<li>Identifying change points: Use similarity scores to detect significant change points in the time series. Change points are typically identified as places where the similarity scores between adjacent segments change abruptly, indicating a change in the behavior of the series.</li>

<li>Segmentation of the time series: Using the identified change points, segment the time series into intervals where the behavior is relatively consistent between consecutive change points.</li>
    </ol>

```python
from claspy.segmentation import BinaryClaSPSegmentation
from claspy.data_loader import load_tssb_dataset
import numpy as np
```

```python
def read_data_from_txt_file(file_path):
    try:
        with open(file_path, 'r') as file:
            lines = file.readlines()
            data = [float(line.strip()) for line in lines if line.strip()]
        xs = np.arange(len(data))
        ys = np.array(data)
        return xs, ys
    except FileNotFoundError:
        print("File not found.")
        return None, None
    except Exception as e:
        print("Error reading data:", e)
        return None, None
```

<h3>Let's first test the ClaSP algorithm to TS data from the "Time Series Segmentation Benchmark" (TSSB) to demonstrate its utility.</h3>

<h3>Adiac dataset</h3>

```python
dataset, window_size, true_cps, time_series = load_tssb_dataset(names=("Adiac",)).iloc[0,:]
clasp = BinaryClaSPSegmentation()
clasp.fit_predict(time_series)
clasp.plot(gt_cps=true_cps, heading="Segmentation of Adiac", ts_name="Adiac")
```
