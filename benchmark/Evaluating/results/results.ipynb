```python
dfHuman = pd.read_csv("./results_human_activity_segmentation.csv")
dfSKAB = pd.read_csv("./Skab.csv")
dfTCPD = pd.read_csv("./TCPD.csv")
dfTSSB = pd.read_csv("./TSSB.csv")
dfMOSAD = pd.read_csv("./MOSAD.csv")


dfs = [('Human_activity', dfHuman), ('SKAB', dfSKAB), ('TCPD', dfTCPD), ('TSSB', dfTSSB),('MOSAD', dfMOSAD)]


rankings = {}


for name, df in dfs:
    
    rankings[name] = {}
    
    for metric in df.columns[1:]:
       
        sorted_df = df.sort_values(by=metric, ascending=False).reset_index(drop=True)
        
        algo_values = [f"{algo} ({value})" for algo, value in zip(sorted_df['name_algo'], sorted_df[metric])]
        
        rankings[name][metric] = algo_values


rankings_df = pd.DataFrame({(dataset, metric): rankings[dataset][metric] 
                            for dataset in rankings 
                            for metric in rankings[dataset]})


rankings_df.to_csv("./algorithm_rankings.csv", index=False)
rankings_df
```
